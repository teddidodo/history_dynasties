# Web Scraping

### To events.csv

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://en.wikipedia.org/wiki/Timeline_of_Korean_history"
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
dynasties = {"1": "Joseon", "2": "Goryeo"}
data_dict = {
    "year": [],
    "description": [],
    "period": [],
    "dynasty_id": []
}
def extract_dynasty(idx, dynasty):
    h2_dynasty = soup.find(id=dynasty)
    country_name = h2_dynasty.text
    ul_dynasty = h2_dynasty.findNext("ul")
    
    for li in ul_dynasty:
        info = li.text.replace("\\n", "").strip().split(":")
        if len(info) == 2:
            year = info[0]
            content = info[1].strip()
            data_dict["year"].append(year)
            data_dict["description"].append(content)
            data_dict["period"].append(str(int(year[:2]) + 1) + " century")
            data_dict["dynasty_id"].append(idx)
```

```python
for dynasty_idx in dynasties:
    extract_dynasty(dynasty_idx,dynasties[dynasty_idx])
```

```python
df = pd.DataFrame(data_dict)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.to_csv("korea.csv")
```

```python
df.to_csv("events.csv", mode="a", header=False)
```

### Joseon

```python
h2_joseon = soup.find(id="Joseon")
country_name = h2_joseon.text
```

```python
h2_joseon
```

```python
ul_joseon = h2_joseon.findNext("ul")
```

```python
COUNTRY = "N/S Korea"
data_dict = {
    "year": [],
    "description": [],
    "period": [],
    "current_country": [],
    "country": []
}
for li in ul_joseon:
    info = li.text.replace("\\n", "").strip().split(":")
    if len(info) == 2:
        year = info[0]
        content = info[1].strip()
        data_dict["year"].append(year)
        data_dict["description"].append(content)
        data_dict["period"].append(str(int(year[:2]) + 1) + " century")
        data_dict["current_country"].append(COUNTRY)
        data_dict["country"].append(country_name)
```

```python
df = pd.DataFrame(data_dict)
```
